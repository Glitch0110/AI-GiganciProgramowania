```python
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz

x = np.linspace(0, 100, 1000)

trimf = fuzz.trimf(x, [20, 50, 80])
trapmf = fuzz.trapmf(x, [10, 30, 70, 90])
gaussmf = fuzz.gaussmf(x, mean=50, sigma=15)

plt.figure(figsize=(12,6))
plt.plot(x, trimf, label="Trjktna (20, 50, 80)", color="orange")
plt.plot(x, trapmf, label="Trapezowa (10, 30, 70, 90)", color="green")
plt.plot(x, gaussmf, label="Gaussowska (mean=50, sigma=15)", color="blue")

plt.title("Funkcje przynalenoci - scikit-fuzzy")
plt.xlabel('x')
plt.ylabel('Przynaleno')
plt.grid()
plt.legend()
plt.ylim(0, 1.1)
plt.show()
```

```python
ciepo = fuzz.trimf(x, [10, 30, 40])
gorco = fuzz.trimf(x, [30, 40, 50])
plt.figure(figsize=(12,6))
plt.plot(x, ciepo, label="Ciepo (10, 30, 40)", color="orange")
plt.plot(x, gorco, label="Gorco (30, 40, 50)", color="green")
plt.title("Funkcje przynalenoci - ciepo i gorco")
plt.xlabel('x')
plt.ylabel('Przynaleno')
plt.grid()
plt.legend()
plt.ylim(0, 1.1)
plt.show()
```
