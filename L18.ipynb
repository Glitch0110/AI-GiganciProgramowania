```python
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz

x = np.linspace(0, 100, 1000)

trimf = fuzz.trimf(x, [20, 50, 80])
trapmf = fuzz.trapmf(x, [10, 30, 70, 90])
gaussmf = fuzz.gaussmf(x, mean=50, sigma=15)

plt.figure(figsize=(12,6))
plt.plot(x, trimf, label="Trjktna (20, 50, 80)", color="orange")
plt.plot(x, trapmf, label="Trapezowa (10, 30, 70, 90)", color="green")
plt.plot(x, gaussmf, label="Gaussowska (mean=50, sigma=15)", color="blue")

plt.title("Funkcje przynalenoci - scikit-fuzzy")
plt.xlabel('x')
plt.ylabel('Przynaleno')
plt.grid()
plt.legend()
plt.ylim(0, 1.1)
plt.show()
```

```python
ciepo = fuzz.trimf(x, [10, 30, 40])
gorco = fuzz.trimf(x, [30, 40, 50])
plt.figure(figsize=(12,6))
plt.plot(x, ciepo, label="Ciepo (10, 30, 40)", color="orange")
plt.plot(x, gorco, label="Gorco (30, 40, 50)", color="red")
plt.title("Funkcje przynalenoci - ciepo i gorco")
plt.xlabel('x')
plt.ylabel('Przynaleno')
plt.grid()
plt.legend()
plt.ylim(0, 1.1)
plt.xlim(0, 60)
plt.show()
```

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

#Zdefiniowanie zmiennych
#wejciowych
life = ctrl.Antecedent(np.linspace(0, 100, 1000), 'ycie gracza')

distance = ctrl.Antecedent(np.linspace(0, 50, 1000), "Odlego")
#wyjciowych
attack = ctrl.Consequent(np.linspace(0,100,1000), 'Poziom ataku')

life['Niskie'] = fuzz.trimf(life.universe, [0, 0, 40])
life['rednie'] = fuzz.trimf(life.universe, [30, 50, 70])
life['Wysokie'] = fuzz.trimf(life.universe, [60, 100, 100])

distance['Maa'] = fuzz.trimf(distance.universe, [0, 0, 15])
distance['Dua'] = fuzz.trimf(distance.universe, [10, 50, 50])

attack['Brak'] = fuzz.trimf(attack.universe, [0, 0, 30])
attack['Ostrony'] = fuzz.trimf(attack.universe, [50, 70, 90])
attack['Zdecydowany'] = fuzz.trimf(attack.universe, [80, 100, 100])

rule1 = ctrl.Rule(life['Niskie'], attack['Zdecydowany'])
rule2 = ctrl.Rule(life["rednie"] & distance['Dua'], attack['Ostrony'])
rule3 = ctrl.Rule(life["rednie"] & distance['Maa'], attack['Ostrony'])
rule4 = ctrl.Rule(life["Wysokie"] & distance['Maa'], attack['Brak'])

attack_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4])
attack_sim = ctrl.ControlSystemSimulation(attack_ctrl)

attack_sim.input['ycie gracza'] = 45

attack_sim.input['Odlego'] = 2

attack_sim.compute()

print(f'Poziom ataku: {attack_sim.output['Poziom ataku']:.2f}')
attack.view(sim=attack_sim)
```
