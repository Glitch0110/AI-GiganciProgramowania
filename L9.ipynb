```python
df = pd.read_excel("Gigathon_Wyniki.xlsx")
df
```

```python
print(df.head(3))
print(df.tail(3))

df["Wynik procentowy"] = df["Punktacja"] / 120 * 100

df
```

```python
df = df.dropna()
df
```

```python
df = df.sort_values(by="Punktacja", ascending=False)
df
```

```python
df["Zakwalikowany"] = df["Wynik procentowy"] > 70
df["Pełnoletność"] = df["Wiek"] >= 18
df
```

```python
df.groupby(["Zakwalikowany", "Wiek"])[["Punktacja", "Wynik procentowy"]].agg(["count", "min", "max", "mean", "median"])
```

```python
df_z = df[df["Zakwalikowany"]]
df_z.to_excel("zakwalifiikowani.xlsx")
```
