```python
elements = [3, 4, 5]
def srednia_arytmetyczna(list:elements):
  suma = 0
  for num in elements:
    suma += num
  return suma / len(elements)
print(srednia_arytmetyczna(elements))

def silnia(num):
  wynik = 1
  for i in range(1, num + 1):
    wynik *= i
  return wynik
print(silnia(3))

def srednia_ważona(elements, weights):
  list_elements = []
  for num in elements:
    for weight in weights:
     list_elements.append(num * weight)
  return srednia_arytmetyczna(list_elements)
print(srednia_ważona([3, 4], [5, 6]))

```

```python
nums = [1, 9, 3, 9, 4]
nums = [1, 3, 4, 9, 9, 19]
def calculate_median(list:nums):
  nums.sort()
  if len(nums) % 2 == 1:
    return nums[len(nums) // 2]
  else:
    return (nums[len(nums) // 2] + nums[len(nums) // 2 - 1]) / 2
print(calculate_median(nums))

#Modalna
list_of_numbers = [4, 2, 7, 2, 8, 2, 9, 7, 7]

def find_mode(list_of_numbers):
  count_occurrence = {i: list_of_numbers.count(i) for i in set(list_of_numbers)}
  mode = [key for value, key  in zip(count_occurrence.values(), count_occurrence.keys()) if value == max(count_occurrence.values())]
  return mode

print(find_mode(list_of_numbers))
```

```python
#Wariancja
import statistics
import matplotlib.pyplot as plt

# Funkcja do obliczania wariancji i średniej
def calculate_variance(group):
    mean = statistics.mean(group)
    variance = sum((x - mean) ** 2 for x in group) / (len(group) - 1)
    return variance, mean

# Dane punktowe dla czterech drużyn
pajaki = [48, 49, 51, 52]
ender_dragony = [0, 0, 100, 100]
szkielety = [10, 30, 70, 90]
zombie = [49, 50, 50, 51]

groups = {
    "Pająki Parkour Masterzy": pajaki,
    "Ender Dragony": ender_dragony,
    "Szkielety Beznogi Snajperzy": szkielety,
    "Zombie Ziomby": zombie
}

# Obliczanie wariancji i średniej dla każdej grupy
variances = {}
means = {}

for name, group in groups.items():
    var, mean = calculate_variance(group)
    variances[name] = var
    means[name] = mean
    print(f"{name} - Średnia: {mean}, Wariancja: {var}")

# Tworzenie wykresu
plt.figure(figsize=(10, 6))
markers = ['o', 's', '^', 'D']
colors = ['blue', 'red', 'green', 'purple']

for i, (name, group) in enumerate(groups.items()):
    x_values = list(range(len(group)))
    y_values_mean = [means[name]] * len(group)

    # Scatter plot wartości rzeczywistych
    plt.scatter(x_values, group, color=colors[i], label=f"{name} - Dane rzeczywiste", marker=markers[i])

    # Linia średniej
    plt.plot(x_values, y_values_mean, color=colors[i], linestyle='dashed', label=f"{name} - Średnia")

# Opisy osi i tytuł
plt.xlabel("Indeks wartości")
plt.ylabel("Punkty gracza")
plt.title("Średnia i wartości rzeczywiste punktów dla czterech drużyn Minecraft")

# Legenda
plt.legend(loc='lower right')
plt.show()
```

```python
#Ćwiczenia samodzielne

import statistics
import numpy as np

# --- Zadanie 1: średnia ---
reaction_times = [220, 180, 250, 300, 200, 190]
avg_reaction = statistics.mean(reaction_times)
print("Średni czas reakcji gracza:", avg_reaction)

# --- Zadanie 2: mediana ---
damage = [120, 80, 150, 200, 90, 100, 300]
median_damage = statistics.median(damage)
print("Mediana obrażeń:", median_damage)

# --- Zadanie 3: modalna ---
weapons = ["Karabin", "Snajperka", "Karabin", "Pistolet", "Karabin", "Snajperka", "Pistolet"]
most_used_weapon = statistics.mode(weapons)
print("Najczęściej używana broń:", most_used_weapon)

# --- Zadanie 4: wariancja dwóch kierowców ---
driver1 = [68, 70, 72, 69, 71, 70]
driver2 = [55, 65, 70, 50, 68, 72]

var_driver1 = statistics.variance(driver1)
var_driver2 = statistics.variance(driver2)

print("Wariancja kierowca1:", var_driver1)
print("Wariancja kierowca2:", var_driver2)

if var_driver1 < var_driver2:
    print("Kierowca 1 jest bardziej pewny.")
else:
    print("Kierowca 2 jest bardziej pewny.")

# --- Zadanie 5: korelacja ---
czas_gry = [1, 2, 3, 4, 5, 6, 7, 8]
poziom_szczęścia = [50, 47, 53, 49, 52, 48, 51, 50]

corr = np.corrcoef(czas_gry, poziom_szczęścia)[0, 1]
print("Współczynnik korelacji (czas vs szczęście):", corr)
```
