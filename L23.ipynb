```python
import tensorflow as tf
import numpy as np
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

layer = Dense(units=1, input_shape=[1])
model = Sequential([layer])
model.compile(optimizer='sgd', loss="mean_squared_error")

xs = np.array([30, 40, 50, 60, 70, 80, 90, 100], dtype=int)
ys = np.array([430000, 500000, 750000, 1000000, 1200000, 1300000, 1500000, 1800000],dtype=int)


xs_normalized = []
ys_normalized = []
xs_min = min(xs)
ys_min = min(ys)
xs_max = max(ys)
ys_max = max(ys)
for x in xs:
  xs_normalized.append((x - xs_min)/(xs_max - xs_min))
for y in ys:
  ys_normalized.append((y - ys_min)/(ys_max - ys_min))

model.fit(xs_normalized, ys_normalized, epochs=500)
print(model.predict(np.array([[10.0]])))
weights = layer.get_weights()
print(f"Wz√≥r prostej to y = {weights[0][0]}x + {weights[1][0]}")
```
